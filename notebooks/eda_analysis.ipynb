```python
# eda.ipynb or notebooks/eda_analysis.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load data
df = pd.read_csv('C:/Users/lenovo/aqi-predictor/data/IQ_AIR_DATA.csv')  # adjust path as needed
df.head()
df.info()
df.describe()
df.columns
print(df.columns)


```

```python
plt.figure(figsize=(14,5))
plt.plot(df.index, df['AQI'], color='purple')
plt.title('AQI Over Time')
plt.xlabel('Date')
plt.ylabel('Air Quality Index (AQI)')
plt.grid(True)
plt.show()
```

```python
sns.histplot(df['AQI'], bins=30, kde=True, color='skyblue')
plt.title("AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```python
df['hour'] = df.index.hour
df['day'] = df.index.day_name()
df['month'] = df.index.month
```

```python
sns.boxplot(x='hour', y='AQI', data=df)
plt.title('AQI by Hour of Day')
plt.show()

sns.boxplot(x='day', y='AQI', data=df)
plt.title('AQI by Day of Week')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot AQI over time using the index (Timespace)
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['AQI'], label='AQI', color='green')
plt.ylabel("Air Quality Index (AQI)")
plt.xlabel("Timestamp")
plt.title("AQI Trend Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['AQI'], label='AQI', color='green')

# Highlight high pollution events
high_aqi = df[df['AQI'] > 100]
plt.scatter(high_aqi.index, high_aqi['AQI'], color='red', label='High AQI (>100)', zorder=5)

# Optional annotations (optional: show exact AQI)
for i, row in high_aqi.iterrows():
    plt.annotate(f"{row['AQI']}", (i, row['AQI']), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='red')

plt.title("AQI Trend with High Pollution Events Highlighted")
plt.xlabel("Timestamp")
plt.ylabel("Air Quality Index (AQI)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
